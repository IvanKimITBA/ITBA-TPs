```python
votos_por_partido = df_resultados.groupby('agrupacion_nombre')['votos_cantidad'].sum().sort_values(ascending=False)
total_votos = votos_por_partido.sum()
porcentaje_votos_por_partido = (votos_por_partido / total_votos) * 100

print("Votos por Partido Político:\n", votos_por_partido)
print("\nPorcentaje de Votos por Partido Político:\n", porcentaje_votos_por_partido.round(2).astype(str) + '%')
```

```python
#Cuantos votos obtuvo cada partido politico en la Provincia de Buenos Aires (cantidad total de votos positivos y su porcentaje) para Diputados Nacionales?


df_buenos_aires_diputados = df_resultados[
    (df_resultados['distrito_nombre'] == 'BUENOS AIRES') &
    (df_resultados['cargo_nombre'] == 'DIPUTADO NACIONAL')
]

votos_por_partido_ba_diputados = df_buenos_aires_diputados.groupby('agrupacion_nombre')['votos_cantidad'].sum().sort_values(ascending=False)
total_votos_ba_diputados = votos_por_partido_ba_diputados.sum()
porcentaje_votos_por_partido_ba_diputados = (votos_por_partido_ba_diputados / total_votos_ba_diputados) * 100

print("Votos por Partido Político para Diputados en Buenos Aires:\n", votos_por_partido_ba_diputados)
print("\nPorcentaje de Votos por Partido Político para Diputados en Buenos Aires:\n", porcentaje_votos_por_partido_ba_diputados.round(2).astype(str) + '%')
```

```python
# DICCIONARIO DE COLORES DE PARTIDOS POLÍTICOS
colores_partidos = {
    'ALIANZA LA LIBERTAD AVANZA': '#7D32A8',          # Violeta Institucional
    'ALIANZA FUERZA PATRIA': '#004B93',               # Azul Peronista/Federal
    'FUERZA PATRIA': '#004B93',
    'FRENTE DE IZQUIERDA Y DE TRABAJADORES - UNIDAD': '#ED1C24', # Rojo Izquierda
    'PROPUESTA FEDERAL PARA EL CAMBIO': '#FFD700',    # Amarillo PRO
    'ALIANZA PROVINCIAS UNIDAS': '#009BDB',           # Celeste Federal
    'ALIANZA POTENCIA': '#FF6600',                    # Naranja
    'CIUDADANOS UNIDOS': '#2ECC71',                   # Verde Ciudadano
    'PARTIDO LIBERTARIO': '#000000',                  # Negro (Cba/Nac)
    'DEFENDAMOS CÓRDOBA': '#9B59B6'                   # Morado/Lila
}
```

```python
#GRAFICO #1

# Get the top 5 parties for Buenos Aires
top_5_parties = votos_por_partido_ba_diputados.head(5)
top_5_percentages = porcentaje_votos_por_partido_ba_diputados.head(5)

# Prepare data for plotting
plot_data = pd.DataFrame({
    'Partido Político': top_5_parties.index,
    'Cantidad de Votos': top_5_parties.values,
    'Porcentaje': top_5_percentages.values
})

# Create a horizontal bar chart
plt.figure(figsize=(12, 8))
# Use 'Partido Político' for hue to assign different colors to each bar based on the party

sns.barplot(x='Cantidad de Votos', y='Partido Político', hue='Partido Político', data=plot_data, palette=colores_partidos, legend=False)

plt.title('Top 5 Partidos Políticos para Diputados en Buenos Aires (Votos Positivos)', fontsize=16)
plt.xlabel('Cantidad de Votos', fontsize=12)
plt.ylabel('Partido Político', fontsize=12)

# Add percentages next to the bars
for index, row in plot_data.iterrows():
    plt.text(row['Cantidad de Votos'] + 50000, index, f'{row['Porcentaje']:.2f}%', va='center', ha='left', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#GRAFICO #2

# Filter data for 'CIUDAD AUTÓNOMA DE BUENOS AIRES' and 'DIPUTADO NACIONAL'
df_caba_diputados = df_resultados[
    (df_resultados['distrito_nombre'] == 'CIUDAD AUTÓNOMA DE BUENOS AIRES') &
    (df_resultados['cargo_nombre'] == 'DIPUTADO NACIONAL')
]

# Calculate votes and percentages for CABA Diputados
votos_por_partido_caba_diputados = df_caba_diputados.groupby('agrupacion_nombre')['votos_cantidad'].sum().sort_values(ascending=False)
total_votos_caba_diputados = votos_por_partido_caba_diputados.sum()
porcentaje_votos_por_partido_caba_diputados = (votos_por_partido_caba_diputados / total_votos_caba_diputados) * 100

# Get the top 5 parties for CABA
top_5_parties_caba = votos_por_partido_caba_diputados.head(5)
top_5_percentages_caba = porcentaje_votos_por_partido_caba_diputados.head(5)

# Prepare data for plotting
plot_data_caba = pd.DataFrame({
    'Partido Político': top_5_parties_caba.index,
    'Cantidad de Votos': top_5_parties_caba.values,
    'Porcentaje': top_5_percentages_caba.values
})

# Create a horizontal bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Cantidad de Votos', y='Partido Político', hue='Partido Político', data=plot_data_caba, palette=colores_partidos, legend=False)

plt.title('Top 5 Partidos Políticos para Diputados en Ciudad Autónoma de Buenos Aires (Votos Positivos)', fontsize=16)
plt.xlabel('Cantidad de Votos', fontsize=12)
plt.ylabel('Partido Político', fontsize=12)

# Add percentages next to the bars
for index, row in plot_data_caba.iterrows():
    plt.text(row['Cantidad de Votos'] + 5000, index, f'{row['Porcentaje']:.2f}%', va='center', ha='left', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#GRAFICO #3

# Filter data for 'CÓRDOBA' and 'DIPUTADO NACIONAL'
df_cordoba_diputados = df_resultados[
    (df_resultados['distrito_nombre'] == 'CÓRDOBA') &
    (df_resultados['cargo_nombre'] == 'DIPUTADO NACIONAL')
]

# Calculate votes and percentages for Córdoba Diputados
votos_por_partido_cordoba_diputados = df_cordoba_diputados.groupby('agrupacion_nombre')['votos_cantidad'].sum().sort_values(ascending=False)
total_votos_cordoba_diputados = votos_por_partido_cordoba_diputados.sum()
porcentaje_votos_por_partido_cordoba_diputados = (votos_por_partido_cordoba_diputados / total_votos_cordoba_diputados) * 100

# Get the top 5 parties for Córdoba
top_5_parties_cordoba = votos_por_partido_cordoba_diputados.head(5)
top_5_percentages_cordoba = porcentaje_votos_por_partido_cordoba_diputados.head(5)

# Prepare data for plotting
plot_data_cordoba = pd.DataFrame({
    'Partido Político': top_5_parties_cordoba.index,
    'Cantidad de Votos': top_5_parties_cordoba.values,
    'Porcentaje': top_5_percentages_cordoba.values
})

# Create a horizontal bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Cantidad de Votos', y='Partido Político', hue='Partido Político', data=plot_data_cordoba, palette=colores_partidos, legend=False)

plt.title('Top 5 Partidos Políticos para Diputados en Córdoba (Votos Positivos)', fontsize=16)
plt.xlabel('Cantidad de Votos', fontsize=12)
plt.ylabel('Partido Político', fontsize=12)

# Add percentages next to the bars
for index, row in plot_data_cordoba.iterrows():
    plt.text(row['Cantidad de Votos'] + 5000, index, f'{row['Porcentaje']:.2f}%', va='center', ha='left', fontsize=10)

plt.tight_layout()
plt.show()
```
